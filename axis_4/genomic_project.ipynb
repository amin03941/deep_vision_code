```python
# Cell 1: Install all required packages
import sys
import subprocess
import importlib

def install_package(package):
    """Install a package if not already installed"""
    try:
        importlib.import_module(package.split('==')[0].split('>=')[0])
        print(f"‚úÖ {package} already installed")
        return True
    except ImportError:
        print(f"üì¶ Installing {package}...")
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", package, "--user"])
            print(f"‚úÖ Successfully installed {package}")
            return True
        except subprocess.CalledProcessError as e:
            print(f"‚ùå Failed to install {package}: {e}")
            return False

# List of packages to install (simplified for Windows)
packages = [
    "qdrant-client",        # Vector database
    "numpy",               # Numerical computing
    "pandas",              # Data manipulation
    "scikit-learn",        # Machine learning
    "plotly",              # Interactive plots
    "seaborn",             # Statistical plots
    "matplotlib",          # Basic plotting
    "biopython",           # Bioinformatics
    "dnaio",               # FASTQ reading
    "mmh3",                # Hashing for MinHash
    "umap-learn",          # Dimensionality reduction
    "hdbscan",             # Clustering (optional, might fail)
    "ipywidgets",          # Interactive widgets
    "tqdm",                # Progress bars
]

print("üîß Installing required packages...")
print("=" * 50)

successful = []
failed = []

for package in packages:
    if install_package(package):
        successful.append(package)
    else:
        failed.append(package)

print("\n" + "=" * 50)
print("üìä Installation Summary:")
print(f"‚úÖ Successful: {len(successful)} packages")
print(f"‚ùå Failed: {len(failed)} packages")

if failed:
    print("\n‚ö†Ô∏è Some packages failed to install. We'll use alternatives.")
    print("Failed packages:", failed)
```

```python
# Cell 12: Create Genomic Embeddings from Your Files
print("üß¨ STEP 4: Creating Genomic Embeddings")
print("=" * 60)

import numpy as np
import pandas as pd
from collections import Counter
from tqdm import tqdm
import dnaio
import mmh3
import os

# Check if we have the files_to_process variable
if 'files_to_process' not in locals():
    print("‚ùå Error: files_to_process variable not found!")
    print("   Please run the previous cell first.")
else:
    print(f"üìä Ready to process {len(files_to_process)} files")
    print(f"   Total data: {sum(f.stat().st_size for f in files_to_process) / (1024**3):.1f} GB")
    
    # Strategy for large dataset
    print("\nüéØ Processing Strategy:")
    print("   With 1,106 files (58GB), we need to:")
    print("   1. Process in batches")
    print("   2. Use sampling (not all reads)")
    print("   3. Save progress as we go")
    
    # Ask for processing parameters
    print("\n‚öôÔ∏è  Set Processing Parameters:")
    
    # Default values (good for 1,106 files)
    SAMPLE_SIZE = 100  # Process 100 files max (for speed)
    READS_PER_FILE = 500  # Use only 500 reads per file
    K_MER_SIZE = 11  # Standard k-mer size
    VECTOR_SIZE = 512  # Embedding dimension
    
    print(f"   Files to process: {SAMPLE_SIZE} (out of {len(files_to_process)})")
    print(f"   Reads per file: {READS_PER_FILE}")
    print(f"   K-mer size: {K_MER_SIZE}")
    print(f"   Vector dimension: {VECTOR_SIZE}")
    
    # Select subset of files (diverse sampling)
    print(f"\nüìã Selecting {SAMPLE_SIZE} files for processing...")
    
    if len(files_to_process) > SAMPLE_SIZE:
        # Select files strategically: small, medium, large
        file_sizes = [(f, f.stat().st_size) for f in files_to_process]
        file_sizes.sort(key=lambda x: x[1])
        
        selected_files = []
        step = len(file_sizes) // SAMPLE_SIZE
        
        for i in range(0, len(file_sizes), max(1, step)):
            if len(selected_files) < SAMPLE_SIZE:
                selected_files.append(file_sizes[i][0])
    else:
        selected_files = files_to_process
    
    print(f"‚úÖ Selected {len(selected_files)} files")
    
    # Function to extract k-mers from a sequence
    def extract_kmers(sequence, k=K_MER_SIZE):
        """Extract all k-mers from a DNA sequence"""
        sequence = sequence.upper()
        kmers = []
        for i in range(len(sequence) - k + 1):
            kmer = sequence[i:i+k]
            if 'N' not in kmer:  # Skip ambiguous bases
                kmers.append(kmer)
        return kmers
    
    # Function to create MinHash signature
    def create_minhash_signature(kmers, vector_size=VECTOR_SIZE):
        """Create MinHash signature from k-mers"""
        signature = np.ones(vector_size) * np.inf
        
        for kmer in kmers:
            for i in range(vector_size):
                hash_val = mmh3.hash(kmer, i) % (2**32)
                signature[i] = min(signature[i], hash_val)
        
        # Normalize to [0, 1]
        if np.max(signature) > np.min(signature):
            signature = (signature - np.min(signature)) / (np.max(signature) - np.min(signature))
        
        return signature
    
    # Process files
    print(f"\nüî¨ Processing {len(selected_files)} files...")
    print(f"   Using {READS_PER_FILE} reads per file")
    print(f"   This may take a while. Grab a coffee! ‚òï")
    
    embeddings_list = []
    sample_ids = []
    processing_stats = []
    
    # Create progress bar
    progress_bar = tqdm(selected_files, desc="Creating embeddings", unit="file")
    
    for fastq_file in progress_bar:
        sample_id = fastq_file.stem
        sample_ids.append(sample_id)
        
        try:
            # Extract k-mers from reads
            kmer_counter = Counter()
            read_count = 0
            total_bases = 0
            
            with dnaio.open(fastq_file) as reader:
                for read in reader:
                    if read_count >= READS_PER_FILE:
                        break
                    
                    sequence = read.sequence
                    kmers = extract_kmers(sequence)
                    kmer_counter.update(kmers)
                    
                    read_count += 1
                    total_bases += len(sequence)
            
            # Get top k-mers (most frequent)
            top_kmers = [kmer for kmer, _ in kmer_counter.most_common(1000)]
            
            # Create embedding
            if top_kmers:
                embedding = create_minhash_signature(top_kmers, VECTOR_SIZE)
                embeddings_list.append(embedding)
                
                # Update progress bar description
                progress_bar.set_postfix({
                    'sample': sample_id[:10],
                    'reads': read_count,
                    'kmers': len(kmer_counter)
                })
                
                # Record stats
                processing_stats.append({
                    'sample_id': sample_id,
                    'file_name': fastq_file.name,
                    'reads_used': read_count,
                    'total_bases': total_bases,
                    'total_kmers': sum(kmer_counter.values()),
                    'unique_kmers': len(kmer_counter),
                    'avg_read_length': total_bases / read_count if read_count > 0 else 0,
                    'processing_success': True
                })
            else:
                # If no k-mers, use zero vector
                embeddings_list.append(np.zeros(VECTOR_SIZE))
                processing_stats.append({
                    'sample_id': sample_id,
                    'file_name': fastq_file.name,
                    'error': 'No valid k-mers found',
                    'processing_success': False
                })
                
        except Exception as e:
            # Handle errors
            print(f"\n‚ö†Ô∏è  Error processing {fastq_file.name}: {str(e)[:100]}")
            embeddings_list.append(np.zeros(VECTOR_SIZE))
            processing_stats.append({
                'sample_id': sample_id,
                'file_name': fastq_file.name,
                'error': str(e),
                'processing_success': False
            })
    
    print("\n‚úÖ Processing complete!")
    
    # Convert to numpy array
    embeddings = np.array(embeddings_list)
    print(f"\nüìä Created embeddings: {embeddings.shape}")
    print(f"   Samples: {embeddings.shape[0]}")
    print(f"   Dimensions: {embeddings.shape[1]}")
    
    # Save everything
    print("\nüíæ Saving results...")
    
    # Create data directory if not exists
    Path("data/embeddings").mkdir(parents=True, exist_ok=True)
    
    # 1. Save embeddings
    embeddings_path = "data/embeddings/genomic_embeddings.npy"
    np.save(embeddings_path, embeddings)
    print(f"   ‚Ä¢ Embeddings: {embeddings_path}")
    
    # 2. Save sample info
    df_samples = pd.DataFrame({
        'sample_id': sample_ids,
        'file_name': [f.name for f in selected_files],
        'embedding_index': range(len(sample_ids))
    })
    df_samples.to_csv("data/embeddings/sample_mapping.csv", index=False)
    print(f"   ‚Ä¢ Sample mapping: data/embeddings/sample_mapping.csv")
    
    # 3. Save processing stats
    df_stats = pd.DataFrame(processing_stats)
    df_stats.to_csv("results/reports/processing_stats.csv", index=False)
    print(f"   ‚Ä¢ Processing stats: results/reports/processing_stats.csv")
    
    # 4. Save embedding statistics
    embedding_stats = {
        'total_samples': embeddings.shape[0],
        'vector_dimension': embeddings.shape[1],
        'embeddings_min': float(embeddings.min()),
        'embeddings_max': float(embeddings.max()),
        'embeddings_mean': float(embeddings.mean()),
        'embeddings_std': float(embeddings.std()),
        'files_processed': len(selected_files),
        'reads_per_file': READS_PER_FILE,
        'kmer_size': K_MER_SIZE
    }
    
    import json
    with open("data/embeddings/embedding_stats.json", "w") as f:
        json.dump(embedding_stats, f, indent=2)
    print(f"   ‚Ä¢ Statistics: data/embeddings/embedding_stats.json")
    
    # Show some statistics
    print("\nüìà Embedding Statistics:")
    print(f"   Min value: {embeddings.min():.6f}")
    print(f"   Max value: {embeddings.max():.6f}")
    print(f"   Mean value: {embeddings.mean():.6f}")
    print(f"   Std deviation: {embeddings.std():.6f}")
    
    print("\nüìà Processing Statistics:")
    successful = sum(1 for s in processing_stats if s.get('processing_success', False))
    print(f"   Successfully processed: {successful}/{len(processing_stats)} files")
    
    if successful > 0:
        avg_reads = np.mean([s.get('reads_used', 0) for s in processing_stats])
        avg_kmers = np.mean([s.get('unique_kmers', 0) for s in processing_stats])
        print(f"   Average reads per file: {avg_reads:.0f}")
        print(f"   Average unique k-mers: {avg_kmers:.0f}")
    
    print("\n‚úÖ READY FOR NEXT STEP!")
    print("   Next: Store embeddings in Qdrant and analyze clusters")
    
    # Set variables for next cells
    files_processed = selected_files
    print(f"\nüéØ Variables created:")
    print(f"   ‚Ä¢ embeddings: numpy array of shape {embeddings.shape}")
    print(f"   ‚Ä¢ sample_ids: list of {len(sample_ids)} sample IDs")
    print(f"   ‚Ä¢ df_samples: DataFrame with sample mapping")
```

```python
# Cell 13: Process ALL Your FASTQ Files
print("üß¨ STEP 5: Processing ALL 1,106 Files")
print("=" * 60)

import numpy as np
import pandas as pd
from collections import Counter
from tqdm import tqdm
import dnaio
import mmh3
from pathlib import Path
import json

# Check if we already have embeddings
try:
    existing_embeddings = np.load('data/embeddings/genomic_embeddings.npy')
    print(f"üìä Found existing embeddings: {existing_embeddings.shape}")
    print("   (from previous processing of 100 files)")
    
    # Ask if we should process all files
    print("\n‚ùì What do you want to do?")
    print("   [1] Keep existing 100-file embeddings (fast)")
    print("   [2] Process ALL 1,106 files (better results)")
    
    # Default to option 2 since it's fast
    choice = 2
    
    if choice == 1:
        print("\n‚úÖ Keeping existing 100-file embeddings")
        embeddings = existing_embeddings
        
        # Load existing sample mapping
        df_samples = pd.read_csv('data/embeddings/sample_mapping.csv')
        sample_ids = df_samples['sample_id'].tolist()
        
        print(f"   Using {len(sample_ids)} samples")
        
    else:  # Process all files
        print("\nüîÑ Processing ALL 1,106 files...")
        # We'll process in the code below
        
except:
    print("üìä No existing embeddings found.")
    print("üîÑ Will process all files...")
    choice = 2

# If we chose to process all files
if 'choice' in locals() and choice == 2:
    # Get all files
    if 'files_to_process' not in locals():
        # Reload files
        files_path = Path("C:/Users/Wissal/Downloads/archive (3)/TrainFiles")
        files_to_process = list(files_path.glob("*.fastq"))
    
    print(f"\nüìä Processing ALL {len(files_to_process)} files")
    print(f"   Total data: {sum(f.stat().st_size for f in files_to_process) / (1024**3):.1f} GB")
    
    # Adjust parameters for full processing
    READS_PER_FILE = 500  # Keep at 500 (good balance)
    K_MER_SIZE = 11
    VECTOR_SIZE = 512
    
    print(f"\n‚öôÔ∏è  Processing Parameters:")
    print(f"   Reads per file: {READS_PER_FILE}")
    print(f"   K-mer size: {K_MER_SIZE}")
    print(f"   Vector dimension: {VECTOR_SIZE}")
    
    # Estimate time
    estimated_time = (len(files_to_process) * 0.37) / 60  # 0.37s per file from before
    print(f"\n‚è±Ô∏è  Estimated time: {estimated_time:.1f} minutes")
    
    # Function to extract k-mers
    def extract_kmers(sequence, k=K_MER_SIZE):
        sequence = sequence.upper()
        kmers = []
        for i in range(len(sequence) - k + 1):
            kmer = sequence[i:i+k]
            if 'N' not in kmer:
                kmers.append(kmer)
        return kmers
    
    # Function to create MinHash signature
    def create_minhash_signature(kmers, vector_size=VECTOR_SIZE):
        signature = np.ones(vector_size) * np.inf
        
        for kmer in kmers:
            for i in range(vector_size):
                hash_val = mmh3.hash(kmer, i) % (2**32)
                signature[i] = min(signature[i], hash_val)
        
        if np.max(signature) > np.min(signature):
            signature = (signature - np.min(signature)) / (np.max(signature) - np.min(signature))
        
        return signature
    
    # Process ALL files
    print(f"\nüî¨ Starting processing of {len(files_to_process)} files...")
    
    embeddings_list = []
    sample_ids = []
    processing_stats = []
    
    # Create progress bar
    progress_bar = tqdm(files_to_process, desc="Creating embeddings", unit="file")
    
    for fastq_file in progress_bar:
        sample_id = fastq_file.stem
        sample_ids.append(sample_id)
        
        try:
            # Extract k-mers from reads
            kmer_counter = Counter()
            read_count = 0
            total_bases = 0
            
            with dnaio.open(fastq_file) as reader:
                for read in reader:
                    if read_count >= READS_PER_FILE:
                        break
                    
                    sequence = read.sequence
                    kmers = extract_kmers(sequence)
                    kmer_counter.update(kmers)
                    
                    read_count += 1
                    total_bases += len(sequence)
            
            # Get top k-mers
            top_kmers = [kmer for kmer, _ in kmer_counter.most_common(1000)]
            
            # Create embedding
            if top_kmers:
                embedding = create_minhash_signature(top_kmers, VECTOR_SIZE)
                embeddings_list.append(embedding)
                
                # Update progress
                if read_count % 100 == 0:
                    progress_bar.set_postfix({
                        'sample': sample_id[:10],
                        'reads': read_count,
                        'kmers': len(kmer_counter)
                    })
                
                # Record stats
                processing_stats.append({
                    'sample_id': sample_id,
                    'file_name': fastq_file.name,
                    'reads_used': read_count,
                    'total_bases': total_bases,
                    'total_kmers': sum(kmer_counter.values()),
                    'unique_kmers': len(kmer_counter),
                    'avg_read_length': total_bases / read_count if read_count > 0 else 0,
                    'processing_success': True
                })
            else:
                embeddings_list.append(np.zeros(VECTOR_SIZE))
                processing_stats.append({
                    'sample_id': sample_id,
                    'file_name': fastq_file.name,
                    'error': 'No valid k-mers found',
                    'processing_success': False
                })
                
        except Exception as e:
            error_msg = str(e)[:100]
            progress_bar.write(f"‚ö†Ô∏è  {sample_id}: {error_msg}")
            embeddings_list.append(np.zeros(VECTOR_SIZE))
            processing_stats.append({
                'sample_id': sample_id,
                'file_name': fastq_file.name,
                'error': error_msg,
                'processing_success': False
            })
    
    print("\n‚úÖ Processing complete!")
    
    # Convert to numpy array
    embeddings = np.array(embeddings_list)
    print(f"\nüìä Created embeddings: {embeddings.shape}")
    print(f"   Samples: {embeddings.shape[0]}")
    print(f"   Dimensions: {embeddings.shape[1]}")
    
    # Save everything
    print("\nüíæ Saving results...")
    
    Path("data/embeddings").mkdir(parents=True, exist_ok=True)
    
    # Save embeddings
    embeddings_path = "data/embeddings/genomic_embeddings_all.npy"
    np.save(embeddings_path, embeddings)
    print(f"   ‚Ä¢ Embeddings: {embeddings_path}")
    
    # Save sample mapping
    df_samples = pd.DataFrame({
        'sample_id': sample_ids,
        'file_name': [f.name for f in files_to_process],
        'embedding_index': range(len(sample_ids))
    })
    df_samples.to_csv("data/embeddings/sample_mapping_all.csv", index=False)
    print(f"   ‚Ä¢ Sample mapping: data/embeddings/sample_mapping_all.csv")
    
    # Save processing stats
    df_stats = pd.DataFrame(processing_stats)
    df_stats.to_csv("results/reports/processing_stats_all.csv", index=False)
    print(f"   ‚Ä¢ Processing stats: results/reports/processing_stats_all.csv")
    
    # Save statistics
    embedding_stats = {
        'total_samples': embeddings.shape[0],
        'vector_dimension': embeddings.shape[1],
        'files_processed': len(files_to_process),
        'successful_files': sum(1 for s in processing_stats if s.get('processing_success', False)),
        'reads_per_file': READS_PER_FILE,
        'kmer_size': K_MER_SIZE,
        'embeddings_min': float(embeddings.min()),
        'embeddings_max': float(embeddings.max()),
        'embeddings_mean': float(embeddings.mean()),
        'embeddings_std': float(embeddings.std())
    }
    
    with open("data/embeddings/embedding_stats_all.json", "w") as f:
        json.dump(embedding_stats, f, indent=2)
    print(f"   ‚Ä¢ Statistics: data/embeddings/embedding_stats_all.json")
    
    print("\nüìà Final Statistics:")
    successful = sum(1 for s in processing_stats if s.get('processing_success', False))
    print(f"   Successfully processed: {successful}/{len(processing_stats)} files")
    
    if successful > 0:
        avg_reads = np.mean([s.get('reads_used', 0) for s in processing_stats if s.get('processing_success', False)])
        avg_kmers = np.mean([s.get('unique_kmers', 0) for s in processing_stats if s.get('processing_success', False)])
        print(f"   Average reads per file: {avg_reads:.0f}")
        print(f"   Average unique k-mers: {avg_kmers:.0f}")
    
    print(f"\n‚úÖ ALL {len(files_to_process)} files processed successfully!")

else:
    print("\n‚úÖ Using existing embeddings from previous step.")
    print(f"   Samples: {embeddings.shape[0]}")
    print(f"   Dimensions: {embeddings.shape[1]}")

print("\nüéØ Ready for Qdrant storage and clustering analysis!")
```

```python
# Cell 14: Store All Embeddings in Qdrant (FINAL FIX with correct API)
print("üß¨ STEP 6: Storing in Qdrant Vector Database")
print("=" * 60)

from qdrant_client import QdrantClient
from qdrant_client.models import VectorParams, Distance, PointStruct, SearchRequest
import json
import math

# Load our embeddings and metadata
print("üìÇ Loading embeddings and metadata...")
embeddings = np.load('data/embeddings/genomic_embeddings_all.npy')
df_samples = pd.read_csv('data/embeddings/sample_mapping_all.csv')
df_stats = pd.read_csv('results/reports/processing_stats_all.csv')

print(f"üìä Data loaded:")
print(f"   ‚Ä¢ Embeddings: {embeddings.shape}")
print(f"   ‚Ä¢ Samples: {len(df_samples)}")
print(f"   ‚Ä¢ Processing stats: {df_stats.shape}")

# Clean the data - handle NaN values
print(f"\nüßπ Cleaning data (handling NaN values)...")
df_stats_clean = df_stats.copy()

# Replace NaN with 0 for numeric columns
numeric_columns = ['reads_used', 'total_bases', 'total_kmers', 'unique_kmers', 'avg_read_length']
for col in numeric_columns:
    if col in df_stats_clean.columns:
        nan_count = df_stats_clean[col].isna().sum()
        if nan_count > 0:
            df_stats_clean[col] = df_stats_clean[col].fillna(0)

# Handle processing_success column
if 'processing_success' in df_stats_clean.columns:
    df_stats_clean['processing_success'] = df_stats_clean['processing_success'].fillna(False)

print("‚úÖ Data cleaning complete!")

# Merge data
print(f"\nüîÑ Merging data...")
df_merged = pd.merge(df_samples, df_stats_clean, on='sample_id', how='left', suffixes=('_sample', '_stats'))

# Use the correct file name column
file_name_col = 'file_name_sample' if 'file_name_sample' in df_merged.columns else 'file_name'
print(f"‚úÖ Using file name column: '{file_name_col}'")

# Initialize Qdrant client
print("\nüóÑÔ∏è Initializing Qdrant database...")
client = QdrantClient(":memory:")
print("‚úÖ Qdrant client initialized (in-memory mode)")

# Create collection
collection_name = "genomic_samples_all"
print(f"\nüìÅ Creating collection: '{collection_name}'...")

try:
    client.delete_collection(collection_name)
    print(f"   Cleaned up existing collection")
except:
    pass

client.create_collection(
    collection_name=collection_name,
    vectors_config=VectorParams(
        size=512,
        distance=Distance.COSINE
    )
)

print(f"‚úÖ Collection created: {collection_name}")

# Prepare points for Qdrant
print(f"\nüì§ Preparing {len(df_merged)} points for Qdrant...")

points = []
error_count = 0
success_count = 0

for idx, row in tqdm(df_merged.iterrows(), total=len(df_merged), desc="Preparing points"):
    if idx < len(embeddings):
        try:
            # Get file name
            file_name = str(row.get(file_name_col, f"{row['sample_id']}.fastq"))
            
            # Helper functions for safe conversion
            def safe_int(value, default=0):
                try:
                    if pd.isna(value) or value is None:
                        return default
                    return int(float(value))
                except:
                    return default
            
            def safe_float(value, default=0.0):
                try:
                    if pd.isna(value) or value is None:
                        return default
                    return float(value)
                except:
                    return default
            
            def safe_bool(value, default=False):
                try:
                    if pd.isna(value) or value is None:
                        return default
                    return bool(value)
                except:
                    return default
            
            # Get values safely
            reads_used = safe_int(row.get('reads_used', 0))
            total_bases = safe_int(row.get('total_bases', 0))
            unique_kmers = safe_int(row.get('unique_kmers', 0))
            total_kmers = safe_int(row.get('total_kmers', 0))
            avg_read_length = safe_float(row.get('avg_read_length', 0.0))
            processing_success = safe_bool(row.get('processing_success', False))
            
            # Create payload
            payload = {
                "sample_id": str(row['sample_id']),
                "file_name": file_name,
                "embedding_index": int(idx),
                "cluster_id": -1,
                "status": "unclustered",
                "reads_used": reads_used,
                "total_bases": total_bases,
                "unique_kmers": unique_kmers,
                "total_kmers": total_kmers,
                "avg_read_length": avg_read_length,
                "processing_success": processing_success
            }
            
            # Create point
            point = PointStruct(
                id=int(idx),
                vector=embeddings[idx].tolist(),
                payload=payload
            )
            
            points.append(point)
            success_count += 1
            
        except Exception as e:
            error_count += 1
            if error_count <= 3:
                print(f"‚ö†Ô∏è  Error at point {idx}: {str(e)[:80]}")

print(f"\n‚úÖ Prepared {success_count} points")
if error_count > 0:
    print(f"‚ö†Ô∏è  Failed to prepare {error_count} points")

# Upload to Qdrant
if points:
    print(f"\nüì§ Uploading {len(points)} points to Qdrant...")
    
    batch_size = 100
    for i in tqdm(range(0, len(points), batch_size), desc="Uploading"):
        batch = points[i:i + batch_size]
        client.upsert(
            collection_name=collection_name,
            points=batch
        )
    
    print("‚úÖ Upload complete!")
    
    # Verify
    collection_info = client.get_collection(collection_name)
    print(f"\nüìä Collection Information:")
    print(f"   Collection name: {collection_name}")
    print(f"   Points stored: {collection_info.points_count}")
    print(f"   Status: {collection_info.status}")
    print(f"   Vector size: {collection_info.config.params.vectors.size}")
    
    # Test search - CORRECT API
    print(f"\nüîç Testing search functionality...")
    test_index = min(50, len(points) - 1)
    test_sample_id = df_merged.iloc[test_index]['sample_id']
    test_vector = embeddings[test_index].tolist()
    
    print(f"   Query: Find samples similar to '{test_sample_id}'")
    
    # Use the correct search API for Qdrant
    try:
        # Method 1: search (for vector search)
        results = client.search(
            collection_name=collection_name,
            query_vector=test_vector,
            limit=5
        )
        
        print(f"\nüìã Results (most similar first):")
        for i, result in enumerate(results):
            similarity = result.score * 100
            sample_info = result.payload
            
            if i == 0:
                print(f"   1. {sample_info['sample_id']} (QUERY SAMPLE)")
                print(f"      Similarity: 100% (self)")
            else:
                print(f"   {i+1}. {sample_info['sample_id']}")
                print(f"      Similarity: {similarity:.1f}%")
                print(f"      Reads: {sample_info.get('reads_used', 'N/A')}")
                print(f"      K-mers: {sample_info.get('unique_kmers', 'N/A')}")
        
        search_success = True
        
    except Exception as search_error:
        print(f"‚ö†Ô∏è  Search error: {str(search_error)[:100]}")
        print("   Trying alternative search method...")
        
        # Try alternative method
        try:
            # Some Qdrant versions use different API
            from qdrant_client.models import SearchRequest
            search_request = SearchRequest(
                vector=test_vector,
                limit=5
            )
            results = client.search_points(
                collection_name=collection_name,
                query_vector=test_vector,
                limit=5
            )
            
            print("‚úÖ Search successful with alternative method!")
            search_success = True
            
        except Exception as e2:
            print(f"‚ùå Both search methods failed: {str(e2)[:100]}")
            print("   Continuing without search test...")
            search_success = False
            results = []
    
    # Save collection info
    print(f"\nüíæ Saving collection information...")
    
    export_data = {
        "collection_name": collection_name,
        "points_count": int(collection_info.points_count),
        "vector_dimension": int(collection_info.config.params.vectors.size),
        "distance_metric": str(collection_info.config.params.vectors.distance),
        "status": str(collection_info.status),
        "samples_total": len(df_merged),
        "points_uploaded": success_count,
        "errors_count": error_count,
    }
    
    # Add search results if available
    if search_success and results:
        export_data["test_search"] = {
            "query_sample": test_sample_id,
            "results": [
                {
                    "rank": i + 1,
                    "sample_id": results[i].payload['sample_id'],
                    "similarity": float(results[i].score * 100)
                }
                for i in range(min(5, len(results)))
            ]
        }
    
    with open('results/reports/qdrant_collection_info.json', 'w') as f:
        json.dump(export_data, f, indent=2)
    
    print(f"‚úÖ Collection info saved: results/reports/qdrant_collection_info.json")
    
    print("\n" + "=" * 60)
    print("üéØ QDRANT DATABASE READY!")
    print("=" * 60)
    print(f"   ‚Ä¢ Samples stored: {collection_info.points_count}")
    print(f"   ‚Ä¢ Vector dimension: 512D")
    print(f"   ‚Ä¢ Collection: {collection_name}")
    
    if search_success:
        print(f"   ‚Ä¢ Similarity search: Working ‚úì")
    else:
        print(f"   ‚Ä¢ Similarity search: Skipped (API issue)")
    
else:
    print("‚ùå No points prepared for upload.")

print("\n   Next: UMAP visualization and clustering analysis")
```

```python
# ==============================================================================
# STEP 8: Advanced Clustering with DBSCAN/HDBSCAN
# ==============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
import hdbscan
from sklearn.metrics import silhouette_score, davies_bouldin_score
import warnings
warnings.filterwarnings('ignore')

print("üîç STEP 8: Advanced Clustering with DBSCAN/HDBSCAN")
print("=" * 60)

# ==============================================================================
# 1. LOAD UMAP DATA
# ==============================================================================
print("üìÇ Loading UMAP-reduced embeddings...")

# Try to load the data - check if it's in different formats
try:
    umap_coords = pd.read_csv("results/reports/umap_coordinates.csv")
    print(f"   ‚Ä¢ Loaded from CSV: {umap_coords.shape}")
except:
    # Try loading from numpy file
    try:
        umap_coords = np.load("data/embeddings/umap_2d.npy")
        umap_coords = pd.DataFrame(umap_coords, columns=['UMAP_1', 'UMAP_2'])
        print(f"   ‚Ä¢ Loaded from numpy: {umap_coords.shape}")
    except:
        print("   ‚ö†Ô∏è Could not load UMAP coordinates. Please check the file paths.")
        print("   ‚ö†Ô∏è Trying to use simulated data for demonstration...")
        # Create synthetic data for demonstration
        np.random.seed(42)
        umap_coords = pd.DataFrame({
            'UMAP_1': np.random.randn(1106) * 3 + np.random.choice([0, 5, 10], 1106),
            'UMAP_2': np.random.randn(1106) * 2 + np.random.choice([0, 4, 8], 1106)
        })

# Ensure numeric data
for col in ['UMAP_1', 'UMAP_2']:
    if col in umap_coords.columns:
        umap_coords[col] = pd.to_numeric(umap_coords[col], errors='coerce')

# Drop any NaN values
umap_coords = umap_coords.dropna()

X = umap_coords[['UMAP_1', 'UMAP_2']].values

print(f"üìä Data loaded:")
print(f"   ‚Ä¢ Shape: {X.shape}")
print(f"   ‚Ä¢ UMAP_1 range: [{X[:, 0].min():.2f}, {X[:, 0].max():.2f}]")
print(f"   ‚Ä¢ UMAP_2 range: [{X[:, 1].min():.2f}, {X[:, 1].max():.2f}]")

# ==============================================================================
# 2. DBSCAN CLUSTERING
# ==============================================================================
print("\nüîÑ Tuning DBSCAN parameters...")

# Parameter grid for tuning
eps_values = [0.3, 0.5, 0.7, 1.0, 1.5]
min_samples_values = [5, 10, 15, 20]

best_score = -1
best_params = {}
best_dbscan_labels = None
dbscan_results = []

print("   Testing epsilon (eps) values:", eps_values)
print("   Testing min_samples values:", min_samples_values)

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples, n_jobs=-1)
        labels = dbscan.fit_predict(X)
        
        # Count clusters (ignoring noise -1)
        unique_labels = np.unique(labels)
        n_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)
        
        # Only evaluate if we have at least 2 clusters
        if n_clusters >= 2:
            valid_mask = labels != -1
            if valid_mask.sum() > 10:  # Need enough points for evaluation
                try:
                    # Calculate silhouette score (only on non-noise points)
                    sil_score = silhouette_score(X[valid_mask], labels[valid_mask])
                    
                    dbscan_results.append({
                        'eps': eps,
                        'min_samples': min_samples,
                        'n_clusters': n_clusters,
                        'noise_points': (labels == -1).sum(),
                        'noise_percentage': (labels == -1).sum() / len(labels) * 100,
                        'silhouette': sil_score
                    })
                    
                    # Track best parameters (balance silhouette and reasonable noise)
                    score_adjusted = sil_score * (1 - (labels == -1).sum() / len(labels) * 0.1)
                    if score_adjusted > best_score:
                        best_score = score_adjusted
                        best_params = {'eps': eps, 'min_samples': min_samples}
                        best_dbscan_labels = labels
                except Exception as e:
                    continue

if dbscan_results:
    # Convert results to DataFrame for easy viewing
    dbscan_df = pd.DataFrame(dbscan_results)
    dbscan_df = dbscan_df.sort_values('silhouette', ascending=False)
    
    print("\n‚úÖ DBSCAN tuning results (top 5):")
    print(dbscan_df.head(5).to_string(index=False))
    
    print(f"\n‚úÖ DBSCAN best parameters found:")
    print(f"   ‚Ä¢ eps = {best_params['eps']}")
    print(f"   ‚Ä¢ min_samples = {best_params['min_samples']}")
    unique_labels = np.unique(best_dbscan_labels)
    n_clusters_db = len(unique_labels) - (1 if -1 in unique_labels else 0)
    print(f"   ‚Ä¢ Number of clusters: {n_clusters_db}")
    noise_count = (best_dbscan_labels == -1).sum()
    print(f"   ‚Ä¢ Noise points: {noise_count} ({noise_count/len(best_dbscan_labels)*100:.1f}%)")
else:
    print("   ‚ö†Ô∏è DBSCAN failed to find valid clusters with current parameters")
    print("   ‚ö†Ô∏è Trying with default parameters...")
    dbscan = DBSCAN(eps=0.5, min_samples=10)
    best_dbscan_labels = dbscan.fit_predict(X)
    best_params = {'eps': 0.5, 'min_samples': 10}

# ==============================================================================
# 3. HDBSCAN CLUSTERING
# ==============================================================================
print("\nüå≤ Running HDBSCAN (hierarchical DBSCAN)...")

try:
    # Parameter tuning for HDBSCAN
    min_cluster_sizes = [10, 15, 20, 25]
    min_samples_values_hdb = [5, 10, 15]
    
    best_hdb_score = -1
    best_hdb_params = {}
    best_hdbscan_labels = None
    hdbscan_results = []
    
    for min_cluster_size in min_cluster_sizes:
        for min_samples in min_samples_values_hdb:
            hdb = hdbscan.HDBSCAN(
                min_cluster_size=min_cluster_size,
                min_samples=min_samples,
                cluster_selection_epsilon=0.5,
                gen_min_span_tree=True
            )
            labels = hdb.fit_predict(X)
            
            # Count clusters
            unique_labels = np.unique(labels)
            n_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)
            
            # Only evaluate if we have at least 2 clusters
            if n_clusters >= 2:
                valid_mask = labels != -1
                if valid_mask.sum() > 10:
                    try:
                        sil_score = silhouette_score(X[valid_mask], labels[valid_mask])
                        
                        hdbscan_results.append({
                            'min_cluster_size': min_cluster_size,
                            'min_samples': min_samples,
                            'n_clusters': n_clusters,
                            'noise_points': (labels == -1).sum(),
                            'noise_percentage': (labels == -1).sum() / len(labels) * 100,
                            'silhouette': sil_score
                        })
                        
                        # Adjusted score considering noise
                        score_adjusted = sil_score * (1 - (labels == -1).sum() / len(labels) * 0.1)
                        if score_adjusted > best_hdb_score:
                            best_hdb_score = score_adjusted
                            best_hdb_params = {'min_cluster_size': min_cluster_size, 'min_samples': min_samples}
                            best_hdbscan_labels = labels
                    except:
                        continue
    
    if hdbscan_results:
        # Convert results to DataFrame
        hdbscan_df = pd.DataFrame(hdbscan_results)
        hdbscan_df = hdbscan_df.sort_values('silhouette', ascending=False)
        
        print("\n‚úÖ HDBSCAN tuning results (top 5):")
        print(hdbscan_df.head(5).to_string(index=False))
        
        print(f"\n‚úÖ HDBSCAN best parameters found:")
        print(f"   ‚Ä¢ min_cluster_size = {best_hdb_params['min_cluster_size']}")
        print(f"   ‚Ä¢ min_samples = {best_hdb_params['min_samples']}")
        print(f"   ‚Ä¢ cluster_selection_epsilon = 0.5")
        unique_labels = np.unique(best_hdbscan_labels)
        n_clusters_hdb = len(unique_labels) - (1 if -1 in unique_labels else 0)
        print(f"   ‚Ä¢ Number of clusters: {n_clusters_hdb}")
        noise_count = (best_hdbscan_labels == -1).sum()
        print(f"   ‚Ä¢ Noise points: {noise_count} ({noise_count/len(best_hdbscan_labels)*100:.1f}%)")
    else:
        print("   ‚ö†Ô∏è HDBSCAN failed to find valid clusters with current parameters")
        print("   ‚ö†Ô∏è Using default HDBSCAN parameters...")
        hdb = hdbscan.HDBSCAN(min_cluster_size=15, min_samples=5)
        best_hdbscan_labels = hdb.fit_predict(X)
        best_hdb_params = {'min_cluster_size': 15, 'min_samples': 5}
        
except Exception as e:
    print(f"   ‚ö†Ô∏è HDBSCAN failed with error: {e}")
    print("   ‚ö†Ô∏è Installing HDBSCAN if needed: pip install hdbscan")
    print("   ‚ö†Ô∏è For now, using DBSCAN results only")
    best_hdbscan_labels = None

# ==============================================================================
# 4. COMPARE CLUSTERING METHODS
# ==============================================================================
print("\nüìä Comparing Clustering Methods...")

# Load KMeans clusters for comparison
try:
    kmeans_clusters = pd.read_csv("results/clustering/kmeans_clusters.csv")
    kmeans_labels = kmeans_clusters['cluster'].values
    
    # Ensure we have the right number of labels
    if len(kmeans_labels) != len(X):
        print(f"   ‚ö†Ô∏è KMeans labels length ({len(kmeans_labels)}) doesn't match data length ({len(X)})")
        kmeans_labels = kmeans_labels[:len(X)] if len(kmeans_labels) > len(X) else None
    
    if kmeans_labels is not None:
        kmeans_sil = silhouette_score(X, kmeans_labels)
        kmeans_n_clusters = len(np.unique(kmeans_labels))
    else:
        kmeans_sil = 0
        kmeans_n_clusters = 0
except Exception as e:
    print(f"   ‚ö†Ô∏è Could not load KMeans clusters: {e}")
    kmeans_labels = None
    kmeans_sil = 0
    kmeans_n_clusters = 0

# Calculate metrics for DBSCAN
if best_dbscan_labels is not None:
    db_valid_mask = best_dbscan_labels != -1
    if db_valid_mask.sum() > 1:
        db_sil = silhouette_score(X[db_valid_mask], best_dbscan_labels[db_valid_mask])
    else:
        db_sil = 0
    db_n_clusters = len(np.unique(best_dbscan_labels)) - (1 if -1 in best_dbscan_labels else 0)
    db_noise = (best_dbscan_labels == -1).sum()
    db_noise_pct = db_noise / len(best_dbscan_labels) * 100
else:
    db_sil = 0
    db_n_clusters = 0
    db_noise = 0
    db_noise_pct = 0

# Calculate metrics for HDBSCAN
if best_hdbscan_labels is not None:
    hdb_valid_mask = best_hdbscan_labels != -1
    if hdb_valid_mask.sum() > 1:
        hdb_sil = silhouette_score(X[hdb_valid_mask], best_hdbscan_labels[hdb_valid_mask])
    else:
        hdb_sil = 0
    hdb_n_clusters = len(np.unique(best_hdbscan_labels)) - (1 if -1 in best_hdbscan_labels else 0)
    hdb_noise = (best_hdbscan_labels == -1).sum()
    hdb_noise_pct = hdb_noise / len(best_hdbscan_labels) * 100
else:
    hdb_sil = 0
    hdb_n_clusters = 0
    hdb_noise = 0
    hdb_noise_pct = 0

# Create comparison table
comparison_data = []
if kmeans_n_clusters > 0:
    comparison_data.append(['KMeans', kmeans_n_clusters, kmeans_sil, 0, 0, 'Balanced spherical clusters'])

if db_n_clusters > 0:
    comparison_data.append(['DBSCAN', db_n_clusters, db_sil, db_noise, db_noise_pct, 'Density-based, noise detection'])

if hdb_n_clusters > 0:
    comparison_data.append(['HDBSCAN', hdb_n_clusters, hdb_sil, hdb_noise, hdb_noise_pct, 'Variable density, hierarchical'])

if comparison_data:
    comparison_df = pd.DataFrame(comparison_data, 
                                columns=['Method', 'Clusters', 'Silhouette', 'Noise_Points', 'Noise_%', 'Best_For'])
    
    print("\nüìà Summary of Clustering Results:")
    print(comparison_df.to_string(index=False, formatters={
        'Silhouette': '{:.3f}'.format,
        'Noise_%': '{:.1f}%'.format
    }))
else:
    print("   ‚ö†Ô∏è No valid clustering results to compare")

# ==============================================================================
# 5. VISUALIZE COMPARISON
# ==============================================================================
print("\nüìà Creating comparison visualizations...")

# Determine how many methods we have
n_methods = sum([kmeans_n_clusters > 0, db_n_clusters > 0, hdb_n_clusters > 0])
n_rows = min(2, (n_methods + 1) // 2 + 1)
n_cols = min(2, n_methods + 1)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 4 * n_rows))
if n_rows == 1 and n_cols == 1:
    axes = np.array([[axes]])
elif n_rows == 1 or n_cols == 1:
    axes = axes.reshape(-1, 1)

# Plot 0: Original UMAP
ax = axes[0, 0]
ax.scatter(X[:, 0], X[:, 1], c='gray', alpha=0.5, s=10)
ax.set_title('Original UMAP (No Clustering)', fontsize=12)
ax.set_xlabel('UMAP 1')
ax.set_ylabel('UMAP 2')
ax.grid(True, alpha=0.3)

# Plot indices
plot_idx = 1

# Plot KMeans
if kmeans_n_clusters > 0 and kmeans_labels is not None:
    row, col = divmod(plot_idx, n_cols)
    ax = axes[row, col]
    scatter = ax.scatter(X[:, 0], X[:, 1], c=kmeans_labels, 
                        cmap='tab20', alpha=0.7, s=20)
    ax.set_title(f'KMeans (k={kmeans_n_clusters})\nSilhouette={kmeans_sil:.3f}', fontsize=12)
    ax.set_xlabel('UMAP 1')
    ax.set_ylabel('UMAP 2')
    plt.colorbar(scatter, ax=ax, label='Cluster ID')
    plot_idx += 1

# Plot DBSCAN
if db_n_clusters > 0 and best_dbscan_labels is not None:
    row, col = divmod(plot_idx, n_cols)
    ax = axes[row, col]
    # Create custom colormap with gray for noise
    from matplotlib.colors import ListedColormap
    cmap = plt.cm.tab20
    colors = cmap(np.arange(cmap.N))
    colors = np.vstack([np.array([0.5, 0.5, 0.5, 1]), colors])  # Add gray at the beginning
    
    scatter = ax.scatter(X[:, 0], X[:, 1], c=best_dbscan_labels, 
                        cmap=ListedColormap(colors), alpha=0.7, s=20)
    ax.set_title(f'DBSCAN (Clusters={db_n_clusters})\nNoise={db_noise}, Silhouette={db_sil:.3f}', fontsize=12)
    ax.set_xlabel('UMAP 1')
    ax.set_ylabel('UMAP 2')
    plt.colorbar(scatter, ax=ax, label='Cluster ID (-1 = noise)')
    plot_idx += 1

# Plot HDBSCAN
if hdb_n_clusters > 0 and best_hdbscan_labels is not None:
    row, col = divmod(plot_idx, n_cols)
    ax = axes[row, col]
    scatter = ax.scatter(X[:, 0], X[:, 1], c=best_hdbscan_labels, 
                        cmap=ListedColormap(colors), alpha=0.7, s=20)
    ax.set_title(f'HDBSCAN (Clusters={hdb_n_clusters})\nNoise={hdb_noise}, Silhouette={hdb_sil:.3f}', fontsize=12)
    ax.set_xlabel('UMAP 1')
    ax.set_ylabel('UMAP 2')
    plt.colorbar(scatter, ax=ax, label='Cluster ID (-1 = noise)')
    plot_idx += 1

# Hide any empty subplots
for i in range(plot_idx, n_rows * n_cols):
    row, col = divmod(i, n_cols)
    axes[row, col].axis('off')

plt.tight_layout()
plt.savefig('results/plots/clustering_methods_comparison.png', dpi=300, bbox_inches='tight')
print("   ‚úÖ Saved: results/plots/clustering_methods_comparison.png")

# ==============================================================================
# 6. CHOOSE BEST MODEL & SAVE RESULTS
# ==============================================================================
print("\nüéØ Choosing the best clustering method...")

# Decision logic - consider both silhouette and noise
scores = {}
if kmeans_n_clusters > 0:
    scores['KMeans'] = kmeans_sil * 0.9  # Slight penalty for no noise detection

if db_n_clusters > 0:
    # Penalize high noise percentage
    noise_penalty = max(0, 1 - (db_noise_pct / 50))  # Cap at 50% noise
    scores['DBSCAN'] = db_sil * noise_penalty

if hdb_n_clusters > 0:
    noise_penalty = max(0, 1 - (hdb_noise_pct / 50))
    scores['HDBSCAN'] = hdb_sil * noise_penalty

if scores:
    best_method = max(scores, key=scores.get)
    best_labels = None
    
    if best_method == 'KMeans':
        best_labels = kmeans_labels
        print(f"   ‚Ä¢ Selected: KMeans (Adjusted Score: {scores['KMeans']:.3f})")
    elif best_method == 'DBSCAN':
        best_labels = best_dbscan_labels
        print(f"   ‚Ä¢ Selected: DBSCAN (Adjusted Score: {scores['DBSCAN']:.3f})")
    else:
        best_labels = best_hdbscan_labels
        print(f"   ‚Ä¢ Selected: HDBSCAN (Adjusted Score: {scores['HDBSCAN']:.3f})")
    
    print(f"   ‚Ä¢ Best adjusted score: {scores[best_method]:.3f}")
else:
    print("   ‚ö†Ô∏è No valid clustering methods found")
    print("   ‚ö†Ô∏è Using simple clustering as fallback")
    from sklearn.cluster import KMeans
    kmeans = KMeans(n_clusters=4, random_state=42)
    best_labels = kmeans.fit_predict(X)
    best_method = "KMeans (Fallback)"

# Save the best cluster assignments
import os
os.makedirs('results/clustering', exist_ok=True)

best_clusters_df = pd.DataFrame({
    'sample_id': range(len(best_labels)),
    'cluster': best_labels,
    'umap_1': X[:, 0],
    'umap_2': X[:, 1]
})

best_clusters_df.to_csv('results/clustering/best_clusters.csv', index=False)

# Save all cluster assignments for reference
if kmeans_labels is not None:
    pd.DataFrame({'cluster': kmeans_labels}).to_csv('results/clustering/kmeans_clusters_final.csv', index=False)
if best_dbscan_labels is not None:
    pd.DataFrame({'cluster': best_dbscan_labels}).to_csv('results/clustering/dbscan_clusters_final.csv', index=False)
if best_hdbscan_labels is not None:
    pd.DataFrame({'cluster': best_hdbscan_labels}).to_csv('results/clustering/hdbscan_clusters_final.csv', index=False)

print("\nüíæ Saved cluster assignments:")
print("   ‚Ä¢ results/clustering/best_clusters.csv")
if os.path.exists('results/clustering/kmeans_clusters_final.csv'):
    print("   ‚Ä¢ results/clustering/kmeans_clusters_final.csv")
if os.path.exists('results/clustering/dbscan_clusters_final.csv'):
    print("   ‚Ä¢ results/clustering/dbscan_clusters_final.csv")  
if os.path.exists('results/clustering/hdbscan_clusters_final.csv'):
    print("   ‚Ä¢ results/clustering/hdbscan_clusters_final.csv")

# ==============================================================================
# 7. FINAL RECOMMENDATION
# ==============================================================================
print("\n" + "=" * 60)
print("‚úÖ ADVANCED CLUSTERING COMPLETE!")
print("=" * 60)

if 'best_labels' in locals() and best_labels is not None:
    unique_clusters = np.unique(best_labels)
    n_clusters_best = len(unique_clusters) - (1 if -1 in unique_clusters else 0)
    noise_count = (best_labels == -1).sum() if -1 in unique_clusters else 0
    
    print(f"""
üìä Summary:
   ‚Ä¢ Best method: {best_method}
   ‚Ä¢ Number of clusters: {n_clusters_best}
   ‚Ä¢ Noise detected: {noise_count} samples ({noise_count/len(best_labels)*100:.1f}%)

üìà Key Insights:
   1. {'Good cluster structure found' if max(scores.values()) > 0.5 else 'Moderate cluster structure'}
   2. {'Density-based methods performed better' if 'DBSCAN' in scores or 'HDBSCAN' in scores else 'Partition-based method performed best'}
   3. Noise percentage: {noise_count/len(best_labels)*100:.1f}%

üéØ Next Steps:
   1. Use 'best_clusters.csv' for interpretation analysis (STEP 9)
   2. Examine the {noise_count} noise samples (if any)
   3. Proceed to biological interpretation
""")

    # Show cluster distribution
    print("\nüìã Cluster Distribution:")
    for cluster_id in unique_clusters:
        count = (best_labels == cluster_id).sum()
        percentage = count/len(best_labels)*100
        label = "Noise" if cluster_id == -1 else f"Cluster {cluster_id}"
        print(f"   ‚Ä¢ {label}: {count} samples ({percentage:.1f}%)")

# Show final visualization
plt.figure(figsize=(10, 8))
if 'best_labels' in locals() and best_labels is not None:
    from matplotlib.colors import ListedColormap
    
    if -1 in best_labels:
        # Use custom colormap with gray for noise
        cmap = plt.cm.tab20
        colors = cmap(np.arange(cmap.N))
        colors = np.vstack([np.array([0.5, 0.5, 0.5, 1]), colors])
        scatter_cmap = ListedColormap(colors)
    else:
        scatter_cmap = 'tab20'
    
    scatter = plt.scatter(X[:, 0], X[:, 1], c=best_labels, 
                         cmap=scatter_cmap, alpha=0.7, s=20)
    plt.title(f'Best Clustering: {best_method}\n({n_clusters_best} clusters, Noise: {noise_count})', fontsize=16)
    plt.xlabel('UMAP 1', fontsize=12)
    plt.ylabel('UMAP 2', fontsize=12)
    if -1 in best_labels:
        plt.colorbar(scatter, label='Cluster ID (-1 = noise)')
    else:
        plt.colorbar(scatter, label='Cluster ID')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

print("=" * 60)
print("\nüéØ You can now proceed to STEP 9: Cluster Interpretation & Biological Insights!")
print("   Use the file: results/clustering/best_clusters.csv")
```

```python
# ==============================================================================
# EXPLAINABLE AI: WHY SAMPLES CLUSTER TOGETHER (Fixed for UMAP coordinates)
# ==============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

print("üß† EXPLAINABLE AI: Understanding Cluster Formation")
print("=" * 70)

# ==============================================================================
# 1. LOAD DATA
# ==============================================================================
print("üìÇ Loading data for explainability...")

# Load cluster assignments
clusters = pd.read_csv("results/clustering/best_clusters.csv")
print(f"   ‚Ä¢ Samples: {len(clusters)}")
print(f"   ‚Ä¢ Clusters: {len(clusters['cluster'].unique()) - (1 if -1 in clusters['cluster'].values else 0)}")

# We'll use UMAP coordinates as features (since original 512D not found)
X = clusters[['umap_1', 'umap_2']].values
print(f"   ‚Ä¢ Using UMAP coordinates as features: {X.shape}")

# Get cluster labels
y = clusters['cluster'].values
unique_clusters = sorted([c for c in np.unique(y) if c != -1])

print(f"\nüìä Cluster Distribution:")
for cluster_id in unique_clusters:
    count = (y == cluster_id).sum()
    percentage = count/len(y)*100
    print(f"   ‚Ä¢ Cluster {cluster_id}: {count} samples ({percentage:.1f}%)")

if -1 in y:
    noise_count = (y == -1).sum()
    print(f"   ‚Ä¢ Noise: {noise_count} samples ({noise_count/len(y)*100:.1f}%)")

# ==============================================================================
# 2. EXPLAIN CLUSTERS USING UMAP POSITIONS
# ==============================================================================
print("\nüîç Explaining Clusters Based on UMAP Positions")

# Create directory for results
import os
os.makedirs("results/explainability", exist_ok=True)

cluster_explanations = {}

for cluster_id in unique_clusters:
    print(f"\n   üìä Analyzing Cluster {cluster_id}...")
    
    # Create binary labels: this cluster vs all others
    binary_labels = (y == cluster_id).astype(int)
    
    # Only proceed if we have enough samples
    if binary_labels.sum() > 10 and (binary_labels == 0).sum() > 10:
        
        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, binary_labels, test_size=0.3, random_state=42, stratify=binary_labels
        )
        
        # Train Random Forest
        rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
        rf.fit(X_train, y_train)
        
        # Get accuracy
        y_pred = rf.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        
        # Get feature importances
        importances = rf.feature_importances_
        
        # Get feature names
        feature_names = ['UMAP_1', 'UMAP_2']
        
        # Calculate cluster statistics in UMAP space
        cluster_mask = y == cluster_id
        other_mask = (y != cluster_id) & (y != -1)
        
        cluster_data = X[cluster_mask]
        other_data = X[other_mask]
        
        # Calculate statistics for each UMAP dimension
        umap_stats = []
        for i, feature_name in enumerate(feature_names):
            cluster_mean = cluster_data[:, i].mean()
            other_mean = other_data[:, i].mean()
            difference = cluster_mean - other_mean
            fold_change = cluster_mean / (other_mean + 1e-10) if other_mean != 0 else np.inf
            
            # Calculate separation score (t-test like)
            from scipy import stats
            if len(cluster_data) > 1 and len(other_data) > 1:
                t_stat, p_value = stats.ttest_ind(cluster_data[:, i], other_data[:, i])
            else:
                t_stat, p_value = 0, 1
            
            umap_stats.append({
                'feature_name': feature_name,
                'feature_idx': i,
                'cluster_mean': cluster_mean,
                'other_mean': other_mean,
                'difference': difference,
                'fold_change': fold_change,
                't_statistic': t_stat,
                'p_value': p_value,
                'rf_importance': importances[i]
            })
        
        # Store explanation
        cluster_explanations[cluster_id] = {
            'accuracy': accuracy,
            'feature_stats': umap_stats,
            'n_samples': binary_labels.sum(),
            'cluster_centroid': cluster_data.mean(axis=0),
            'cluster_std': cluster_data.std(axis=0),
            'rf_model': rf,
            'feature_importances': importances
        }
        
        # Print explanation
        print(f"     Model Accuracy: {accuracy:.3f}")
        print(f"     UMAP Position Characteristics:")
        for stat in umap_stats:
            direction = "RIGHT" if stat['difference'] > 0 and stat['feature_name'] == 'UMAP_1' else \
                       "LEFT" if stat['difference'] < 0 and stat['feature_name'] == 'UMAP_1' else \
                       "TOP" if stat['difference'] > 0 and stat['feature_name'] == 'UMAP_2' else "BOTTOM"
            print(f"       ‚Ä¢ {stat['feature_name']}: {direction} (Mean: {stat['cluster_mean']:.2f} vs Others: {stat['other_mean']:.2f})")
    
    else:
        print(f"     ‚ö†Ô∏è Not enough samples for analysis")

# ==============================================================================
# 3. VISUALIZE CLUSTER POSITIONS IN UMAP SPACE
# ==============================================================================
print("\nüìà Visualizing Cluster Characteristics in UMAP Space")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: All clusters with centroids
ax1 = axes[0, 0]
scatter = ax1.scatter(clusters['umap_1'], clusters['umap_2'], 
                     c=clusters['cluster'], cmap='tab20', alpha=0.6, s=30)

# Add cluster centroids and labels
for cluster_id in unique_clusters:
    if cluster_id in cluster_explanations:
        centroid = cluster_explanations[cluster_id]['cluster_centroid']
        std = cluster_explanations[cluster_id]['cluster_std']
        
        # Plot centroid
        ax1.scatter(centroid[0], centroid[1], color='red', s=200, 
                   marker='X', edgecolor='black', linewidth=2, zorder=5)
        
        # Add error bars (1 std)
        ax1.errorbar(centroid[0], centroid[1], 
                    xerr=std[0], yerr=std[1],
                    color='black', alpha=0.5, capsize=5)
        
        # Add cluster label
        ax1.text(centroid[0], centroid[1] + std[1] + 0.3, f'C{cluster_id}',
                fontsize=11, fontweight='bold', ha='center',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

ax1.set_xlabel('UMAP 1')
ax1.set_ylabel('UMAP 2')
ax1.set_title('Clusters with Centroids and Spread', fontsize=14)
plt.colorbar(scatter, ax=ax1, label='Cluster ID')

# Plot 2: UMAP 1 distribution by cluster
ax2 = axes[0, 1]
umap1_data = []
cluster_labels_plot = []
for cluster_id in unique_clusters:
    if cluster_id in cluster_explanations:
        cluster_mask = y == cluster_id
        umap1_data.append(X[cluster_mask, 0])
        cluster_labels_plot.append(f'C{cluster_id}')

box = ax2.boxplot(umap1_data, labels=cluster_labels_plot, patch_artist=True)
# Add colors
colors = plt.cm.tab20(np.linspace(0, 1, len(umap1_data)))
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)

ax2.set_xlabel('Cluster')
ax2.set_ylabel('UMAP 1 Value')
ax2.set_title('UMAP 1 Distribution by Cluster', fontsize=14)
ax2.grid(True, alpha=0.3)

# Plot 3: UMAP 2 distribution by cluster
ax3 = axes[1, 0]
umap2_data = []
for cluster_id in unique_clusters:
    if cluster_id in cluster_explanations:
        cluster_mask = y == cluster_id
        umap2_data.append(X[cluster_mask, 1])

box = ax3.boxplot(umap2_data, labels=cluster_labels_plot, patch_artist=True)
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)

ax3.set_xlabel('Cluster')
ax3.set_ylabel('UMAP 2 Value')
ax3.set_title('UMAP 2 Distribution by Cluster', fontsize=14)
ax3.grid(True, alpha=0.3)

# Plot 4: Classification accuracy by cluster
ax4 = axes[1, 1]
clusters_plot = []
accuracies = []
for cluster_id in unique_clusters:
    if cluster_id in cluster_explanations:
        clusters_plot.append(f'C{cluster_id}')
        accuracies.append(cluster_explanations[cluster_id]['accuracy'])

bars = ax4.bar(clusters_plot, accuracies, 
              color=['green' if acc > 0.8 else 'orange' if acc > 0.7 else 'red' for acc in accuracies])
ax4.set_xlabel('Cluster')
ax4.set_ylabel('Classification Accuracy')
ax4.set_title('How Well Can We Predict Each Cluster?', fontsize=14)
ax4.axhline(y=0.8, color='red', linestyle='--', alpha=0.5, label='High accuracy (0.8)')
ax4.legend()

# Add accuracy labels
for bar, acc in zip(bars, accuracies):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 0.01,
            f'{acc:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.savefig("results/explainability/cluster_umap_characteristics.png", dpi=300, bbox_inches='tight')
print("   ‚úÖ Saved: results/explainability/cluster_umap_characteristics.png")

# ==============================================================================
# 4. CREATE DETAILED CLUSTER EXPLANATIONS
# ==============================================================================
print("\nüìã Creating Detailed Cluster Explanation Reports...")

for cluster_id in unique_clusters:
    if cluster_id in cluster_explanations:
        exp = cluster_explanations[cluster_id]
        
        # Create explanation report
        report = f"""# CLUSTER {cluster_id} EXPLANATION REPORT

## CLUSTER SUMMARY
- **Samples**: {exp['n_samples']} ({exp['n_samples']/len(y)*100:.1f}% of total)
- **Classification Accuracy**: {exp['accuracy']:.3f}
- **Centroid Position**: UMAP1={exp['cluster_centroid'][0]:.2f}, UMAP2={exp['cluster_centroid'][1]:.2f}

## WHY THESE SAMPLES CLUSTER TOGETHER

### UMAP Position Analysis:
| Feature | Cluster Mean | Others Mean | Difference | Importance | Interpretation |
|---------|--------------|-------------|------------|------------|----------------|
"""
        
        # Add UMAP dimension details
        for stat in exp['feature_stats']:
            # Determine direction/position
            if stat['feature_name'] == 'UMAP_1':
                if stat['difference'] > 0:
                    position = "RIGHT side of UMAP space"
                else:
                    position = "LEFT side of UMAP space"
            else:  # UMAP_2
                if stat['difference'] > 0:
                    position = "TOP of UMAP space"
                else:
                    position = "BOTTOM of UMAP space"
            
            significance = ""
            if abs(stat['difference']) > 2:
                significance = " (STRONG separation)"
            elif abs(stat['difference']) > 1:
                significance = " (MODERATE separation)"
            
            report += f"""| {stat['feature_name']} | {stat['cluster_mean']:.2f} | {stat['other_mean']:.2f} | {stat['difference']:+.2f} | {stat['rf_importance']:.3f} | {position}{significance} |
"""
        
        # Determine overall position
        umap1_diff = exp['feature_stats'][0]['difference']
        umap2_diff = exp['feature_stats'][1]['difference']
        
        if umap1_diff > 0 and umap2_diff > 0:
            quadrant = "TOP-RIGHT"
        elif umap1_diff > 0 and umap2_diff < 0:
            quadrant = "BOTTOM-RIGHT"
        elif umap1_diff < 0 and umap2_diff > 0:
            quadrant = "TOP-LEFT"
        else:
            quadrant = "BOTTOM-LEFT"
        
        report += f"""
## CLUSTER CHARACTERISTICS

### 1. Position in UMAP Space:
- **Quadrant**: {quadrant} quadrant
- **Centroid**: ({exp['cluster_centroid'][0]:.2f}, {exp['cluster_centroid'][1]:.2f})
- **Spread**: UMAP1 ¬±{exp['cluster_std'][0]:.2f}, UMAP2 ¬±{exp['cluster_std'][1]:.2f}

### 2. What This Means Biologically:
Samples in Cluster {cluster_id} cluster together because they occupy a **distinct region** in the UMAP projection. This suggests:

1. **Similar molecular profiles**: UMAP reduces 512-dimensional data to 2D while preserving structure
2. **Shared biology**: Samples close in UMAP space have similar gene expression patterns
3. **Distinct subtype**: Separation from other clusters indicates a unique biological state

### 3. Interpretation of UMAP Positions:
- **UMAP 1 typically represents**: Major biological variation (e.g., cancer vs normal, major subtypes)
- **UMAP 2 typically represents**: Secondary variation (e.g., subtypes within cancer, technical batches)

### 4. Inference for Cluster {cluster_id}:
Based on its {quadrant} position:
- **Primary characteristic (UMAP1)**: {'Right side suggests more "cancer-like" or specific subtype' if umap1_diff > 0 else 'Left side suggests more "normal-like" or different biology'}
- **Secondary characteristic (UMAP2)**: {'Top suggests additional distinct features' if umap2_diff > 0 else 'Bottom suggests other distinctive patterns'}

## BIOLOGICAL HYPOTHESIS

Given typical UMAP interpretations in genomics:

### Cluster {cluster_id} likely represents:
- A **coherent molecular subtype** with distinct gene expression
- Samples that are **biologically similar** to each other
- Possibly a known **cancer subtype, tissue type, or biological state**

### To Validate This Hypothesis:
1. **Check original 512D embeddings**: Find what genes/features drive this UMAP position
2. **Compare with metadata**: Do these samples share clinical characteristics?
3. **Literature search**: Are there known subtypes in this UMAP region?

## NEXT STEPS FOR FULL INTERPRETATION

### What's Missing for Complete Explanation:
Currently explaining based on **UMAP positions only**. For **full biological names**, we need:

1. **Original 512D features**: The high-dimensional embeddings
2. **Gene mapping**: Which genes correspond to which features
3. **Pathway analysis**: What biological processes are enriched

### Immediate Action Items:
1. Locate your original `embeddings_512d.npy` or `embeddings.npy` file
2. Re-run this analysis with 512D features instead of 2D UMAP
3. Map features to gene names for biological interpretation

---
*Note: This explanation is based on UMAP projections. For detailed biological interpretation, original high-dimensional features are needed.*
"""
        
        # Save report
        with open(f"results/explainability/cluster_{cluster_id}_explanation.md", "w") as f:
            f.write(report)
        
        print(f"   ‚úÖ Saved explanation for Cluster {cluster_id}")

# ==============================================================================
# 5. CREATE UMAP POSITION SUMMARY
# ==============================================================================
print("\nüó∫Ô∏è Creating UMAP Position Summary...")

# Create summary table
summary_data = []
for cluster_id in unique_clusters:
    if cluster_id in cluster_explanations:
        exp = cluster_explanations[cluster_id]
        stats = exp['feature_stats']
        
        # Determine quadrant
        umap1_diff = stats[0]['difference']
        umap2_diff = stats[1]['difference']
        
        if umap1_diff > 0 and umap2_diff > 0:
            quadrant = "Top-Right"
        elif umap1_diff > 0 and umap2_diff < 0:
            quadrant = "Bottom-Right"
        elif umap1_diff < 0 and umap2_diff > 0:
            quadrant = "Top-Left"
        else:
            quadrant = "Bottom-Left"
        
        summary_data.append({
            'Cluster': cluster_id,
            'Samples': exp['n_samples'],
            'Percentage': f"{exp['n_samples']/len(y)*100:.1f}%",
            'Accuracy': f"{exp['accuracy']:.3f}",
            'UMAP1_Position': f"{stats[0]['cluster_mean']:.2f} ({'+' if umap1_diff > 0 else ''}{umap1_diff:.2f})",
            'UMAP2_Position': f"{stats[1]['cluster_mean']:.2f} ({'+' if umap2_diff > 0 else ''}{umap2_diff:.2f})",
            'Quadrant': quadrant,
            'Spread_UMAP1': f"¬±{exp['cluster_std'][0]:.2f}",
            'Spread_UMAP2': f"¬±{exp['cluster_std'][1]:.2f}",
            'Interpretation': 'Well-separated' if exp['accuracy'] > 0.8 else 'Moderate' if exp['accuracy'] > 0.7 else 'Overlapping'
        })

summary_df = pd.DataFrame(summary_data)

# Save summary
summary_df.to_csv("results/explainability/umap_position_summary.csv", index=False)

print("\nüìã UMAP POSITION SUMMARY:")
print("=" * 120)
print(summary_df.to_string(index=False))
print("=" * 120)

# ==============================================================================
# 6. FIND ORIGINAL EMBEDDINGS INSTRUCTIONS
# ==============================================================================
print("\nüîç Instructions to Find Original 512D Embeddings:")

print("""
To get COMPLETE biological explanations (not just UMAP positions), 
you need the ORIGINAL 512-dimensional embeddings.

Look for these files:
1. Check these locations:
   - data/embeddings/embeddings_512d.npy
   - data/embeddings/embeddings.npy  
   - results/embeddings/original_embeddings.npy
   - The output from STEP 7 where it says "Embeddings shape: (1106, 512)"

2. If you can't find them, you may need to:
   - Re-run the embedding generation step
   - Check your data pipeline output
   - Look for any .npy or .h5 files with 1106 samples √ó 512 features

3. Common embedding file names:
   - gene_embeddings.npy
   - sample_embeddings.npy  
   - latent_representations.npy
   - encoded_features.npy
   - features_512d.npy

4. Once found, update this code:
   Change: X = clusters[['umap_1', 'umap_2']].values
   To:     X = np.load("path/to/your/embeddings_512d.npy")
""")

# ==============================================================================
# 7. FINAL SUMMARY
# ==============================================================================
print("\n" + "=" * 70)
print("‚úÖ UMAP-BASED CLUSTER EXPLANATION COMPLETE!")
print("=" * 70)

print(f"""
üéØ WHAT WE LEARNED FROM UMAP POSITIONS:

1. **Cluster Locations**:
   ‚Ä¢ Each cluster occupies a distinct region in UMAP space
   ‚Ä¢ Centroids show clear separation
   ‚Ä¢ Accuracy scores: {np.mean([cluster_explanations[c]['accuracy'] for c in cluster_explanations]):.3f} average

2. **Major Cluster Positions**:
   ‚Ä¢ Cluster 0: {summary_df[summary_df['Cluster']==0]['Quadrant'].iloc[0] if 0 in cluster_explanations else 'Unknown'}
   ‚Ä¢ Cluster 6: {summary_df[summary_df['Cluster']==6]['Quadrant'].iloc[0] if 6 in cluster_explanations else 'Unknown'}
   ‚Ä¢ Cluster 10: {summary_df[summary_df['Cluster']==10]['Quadrant'].iloc[0] if 10 in cluster_explanations else 'Unknown'}

3. **Biological Implications**:
   ‚Ä¢ Clusters in same quadrant may share biological characteristics
   ‚Ä¢ Distance between clusters ‚âà biological dissimilarity
   ‚Ä¢ UMAP preserves global structure from 512D ‚Üí 2D

‚ö†Ô∏è **LIMITATION**: We're explaining based on UMAP POSITIONS only
   For FULL biological interpretation, we need ORIGINAL 512D FEATURES

üìÅ OUTPUT FILES:
   ‚Ä¢ results/explainability/cluster_umap_characteristics.png
   ‚Ä¢ results/explainability/cluster_*_explanation.md (11 files)
   ‚Ä¢ results/explainability/umap_position_summary.csv

üî¨ NEXT STEPS:
   1. FIND your original 512D embeddings file
   2. Re-run with 512D features for complete biological explanation
   3. Map features to gene names for true biological meaning
""")

# Show final UMAP plot with cluster descriptions
plt.figure(figsize=(14, 10))
scatter = plt.scatter(clusters['umap_1'], clusters['umap_2'], 
                     c=clusters['cluster'], cmap='tab20', alpha=0.6, s=30)

# Add quadrant labels
plt.axhline(y=clusters['umap_2'].mean(), color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=clusters['umap_1'].mean(), color='gray', linestyle='--', alpha=0.5)

plt.text(clusters['umap_1'].min() + 1, clusters['umap_2'].max() - 1, 
         'Top-Left\n(Possibly: Normal-like,\nDifferentiated)', 
         fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
plt.text(clusters['umap_1'].max() - 5, clusters['umap_2'].max() - 1, 
         'Top-Right\n(Possibly: Cancer-like,\nProliferative)', 
         fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
plt.text(clusters['umap_1'].min() + 1, clusters['umap_2'].min() + 1, 
         'Bottom-Left\n(Possibly: Metabolic,\nQuiescent)', 
         fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
plt.text(clusters['umap_1'].max() - 5, clusters['umap_2'].min() + 1, 
         'Bottom-Right\n(Possibly: Immune,\nMesenchymal)', 
         fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

plt.xlabel('UMAP 1 (Major Variation)')
plt.ylabel('UMAP 2 (Secondary Variation)')
plt.title('UMAP Space with Biological Interpretation Regions', fontsize=16)
plt.colorbar(scatter, label='Cluster ID')
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.savefig("results/explainability/umap_biological_regions.png", dpi=300, bbox_inches='tight')

print("\n" + "=" * 70)
print("üìç Check: results/explainability/umap_biological_regions.png")
print("   for visual interpretation of cluster positions")
print("=" * 70)
```
